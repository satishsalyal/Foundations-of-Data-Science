# ðŸ“Š Common Matplotlib Graphs and Plots for EDA
This notebook demonstrates various types of plots available in **Matplotlib** that use for EDA. Each section includes explanations and example code for visualisation using a small dataset.

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
data = np.random.randn(1000)
```

## 1. ðŸŸ¦ Line Plot

**Purpose:** Shows trends or changes in data over a continuous range (like time, distance, etc.).

**Axes:** X-axis â†’ independent variable, Y-axis â†’ dependent variable.

**Common Uses:** Stock prices, temperature trends, sensor readings.

```python
plt.plot(x, y, label='sin(x)')
plt.plot(x, z, label='cos(x)')
plt.title('Line Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

## 2. ðŸ”µ Scatter Plot

**Purpose:** Displays the relationship (correlation) between two variables.

**Visual Clues:** Each point represents one observation.

**Highlights:** Outliers, clusters, or correlations (positive/negative).

```python
plt.scatter(x, y, color='orange', alpha=0.7)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## 3. ðŸŸ© Bar Plot

**Purpose:** Compares quantities across categories.

**Orientation:** Vertical bars; each barâ€™s height = category value.

**Highlights:** Differences between groups or items.

```python
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]
plt.bar(categories, values, color='teal')
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

## 4. Horizontal Bar Plot

A **horizontal bar plot** can be useful when category labels are long.

```python
plt.barh(categories, values, color='purple')
plt.title('Horizontal Bar Plot')
plt.xlabel('Value')
plt.ylabel('Category')
plt.show()
```

## 5. ðŸŸ¨ Histogram

**Purpose:** Shows the distribution of numerical data.

**Bins:** Data is grouped into ranges (â€œbinsâ€), and bar height shows frequency.

**Highlights:** Shape of data (normal, skewed, bimodal).

```python
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## 6. ðŸŸ¥ Box Plot

**Purpose:** Summarizes data distribution using five-number summary (min, Q1, median, Q3, max).

**Highlights:** Spread, central tendency, and outliers.

**Ideal for:** Comparing distributions across multiple datasets.

```python
plt.boxplot(data)
plt.title('Box Plot')
plt.ylabel('Value')
plt.show()
```

## 7. ðŸŸª Violin Plot

**Purpose:** Combines a box plot and a kernel density plot.

**Highlights:** Shows both summary statistics and the full distribution shape.

More informative than box plots when data has multiple peaks.

```python
plt.violinplot(data)
plt.title('Violin Plot')
plt.ylabel('Value')
plt.show()
```

## 8. ðŸŸ§ Pie Chart

**Purpose:** Displays proportions or percentages of a whole.

**Visual Clues:** Each slice represents a categoryâ€™s contribution.

**Highlights:** Relative share of each category (best with few categories).


```python
sizes = [30, 25, 20, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.show()
```

## 9. ðŸŸ© Heatmap

**Purpose:** Represents values in a 2D grid using colors.

**Highlights:** Patterns, correlations, or intensity variations.

**Common Uses:** Correlation matrices, confusion matrices, sensor data.

```python
matrix = np.random.rand(10, 10)
plt.imshow(matrix, cmap='viridis')
plt.title('Heatmap')
plt.colorbar(label='Intensity')
plt.show()
```

## 10. ðŸŸ¨ Polar Plot

**Purpose:** Plots data in polar coordinates (angle and radius).

**Highlights:** Cyclical or circular patterns (like wind direction or frequency response).

**Axes:** Î¸ (angle) and r (radius).

```python
theta = np.linspace(0, 2*np.pi, 100)
r = np.abs(np.sin(5*theta))
plt.polar(theta, r)
plt.title('Polar Plot')
plt.show()
```

## 11. Purpose: Adds depth to visualize relationships among three variables (X, Y, Z).

**Highlights:** Surfaces, trajectories, or spatial relationships.

**Requires:** mpl_toolkits.mplot3d.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot3D(x, y, z, 'red')
ax.set_title('3D Line Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

## 12. Subplots

You can create multiple plots in one figure using **subplots**.

```python
fig, axs = plt.subplots(2, 2, figsize=(8,6))
axs[0,0].plot(x, y)
axs[0,0].set_title('Line')
axs[0,1].scatter(x, y)
axs[0,1].set_title('Scatter')
axs[1,0].hist(data, bins=20)
axs[1,0].set_title('Histogram')
axs[1,1].boxplot(data)
axs[1,1].set_title('Boxplot')
plt.tight_layout()
plt.show()
```

## ðŸ”µ Radar Chart (Spider Chart / Web Chart)

**Purpose:** Displays multivariate data in a circular layout â€” each variable has its own axis arranged radially.

**Highlights:** Excellent for comparing profiles or performances across multiple attributes.

Each polygon represents a data series.

Easy to see strengths/weaknesses or balance between variables.

**Axes:** Each axis = a variable; distance from the center = value.

**Common Uses:*** Skill comparisons, performance metrics, survey data.

```python
import matplotlib.pyplot as plt
import numpy as np

# Variables
categories = ['Accuracy', 'Speed', 'Reliability', 'Scalability', 'Security']
values = [80, 60, 90, 70, 85]
N = len(categories)

# Repeat the first value to close the radar chart
values += values[:1]
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]

# Create radar chart
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(angles, values, linewidth=2, linestyle='solid')
ax.fill(angles, values, 'skyblue', alpha=0.4)

# Set category labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
plt.title('Radar Chart Example')
plt.show()
```

## ðŸŸ  Area Chart

**Purpose:** Shows how one or more quantities change over time, similar to a line plot, but the area under the line is filled to emphasize magnitude.

**Highlights:**

Cumulative totals or stacked trends.

Visualizes part-to-whole relationships over time.

Works best when comparing few series with clear differences.

**Axes:** X-axis â†’ time or continuous variable.

Y-axis â†’ value or magnitude.

**Common Uses:** Financial data, population growth, cumulative sums, etc.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 0.1)
y1 = np.sin(x)
y2 = np.sin(x) + 0.5

plt.fill_between(x, y1, color='skyblue', alpha=0.5, label='sin(x)')
plt.fill_between(x, y2, color='orange', alpha=0.5, label='sin(x)+0.5')
plt.title('Area Chart')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
